# PROTEIN DESIGN

In this task you will use a protein design pipeline based on the combination of **RFDiffusion**, a generative model that can diffuse protein like backbone configurations. **ProteinMPNN** that given a backbone conformation can design a sequence that will fold into it. **AlphaFold2** to assess if the resulting protein structure is predicted consistently with the desired design. In this instruction file you will focus on producing an homoligomer, while for the report you will try to produce a de-novo binder as well as scaffolding an existing protein motif into a new protein.

The pipeline is implemented here: https://colab.research.google.com/github/sokrypton/ColabDesign/blob/main/rf/examples/diffusion.ipynb#scrollTo=pZQnHLuDCsZm

## Additional options for the design of a binder

When designing a binder you need to define your target protein, this is done here by providing a PDB identifier (for example 5cs7 for beta-2-microglobulin) and in contigs indicating first which chain(s) to use:
```
contigs: A:70
pdb: 5cs7      
```
The above line means that we want to bind chain A from the pdb 5cs7, and build a binder of 70 aa.


When doing a binder you should do some changes also to the **AF2** step:
```
initial_guess=True
num_recycles=3
use_multimer=True
```

These options allow AF2 to use the PDB structure in input, increase the prediction accuracy and focus on a multimeric interaction.

## Additional options for the design a protein that can scaffold an active site

When you want to design a protein scaffold around an existing protein site you need again to provide the protein and indicate which parts to reuse and what to do:
```
contigs: 35/A25-27/35
pdb: 1hiv
```
The above line means start your protein with 35 AA, then add the region 25-27 from chain A of the 1HIV model, then add 35 more AA.
